```python
#Load Packages
```

```python
import matplotlib
matplotlib.use('Agg')
 
# import the necessary packages
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import img_to_array
from keras.utils import to_categorical
from keras.models import save_model,load_model
from keras.callbacks import EarlyStopping
from LeNet_model import LeNet
from imutils import paths
import matplotlib.pyplot as plt
import numpy as np
import argparse
from random import sample
import cv2
import os
import glob
```

```python
#Definied the parameters of model with epochs, learning rate and batch size
```

```python
EPOCHS = 100
INIT_LR = 1e-3
BS = 32
```

```python
#Create list of data of train and test
#The resized of image will be 55x55
```

```python
imagePaths = sorted(list(paths.list_images('train')))
imagePaths_train = sample(imagePaths,int(len(imagePaths)*0.8))
index_val=np.where(np.isin(np.array(imagePaths),np.array(imagePaths_train))==False)[0]
imagePaths_val=list(np.array(imagePaths)[index_val])

del imagePaths, index_val


data_train = []
label_train = []

data_val = []
label_val= []

for imagePath in imagePaths_train:
    image = cv2.imread(imagePath)
    image = cv2.resize(image, (55, 55))
    image = img_to_array(image)
    data_train.append(image)
    label_train.append(np.where(imagePath.rfind('dog')>=0,1,0))

for imagePath in imagePaths_val:
    image = cv2.imread(imagePath)
    image = cv2.resize(image, (55, 55))
    image = img_to_array(image)
    data_val.append(image)
    label_val.append(np.where(imagePath.rfind('dog')>=0,1,0))
    
```

```python
#Normalized of images train and test
```

```python
data_train = np.array(data_train, dtype="float") / 255.0
label_train = np.array(label_train)

data_val = np.array(data_val, dtype="float") / 255.0
label_val = np.array(label_val)

trainX=data_train.copy()
valX=data_val.copy()

del data_train, data_val
```

```python
#Read model
```

```python
trainY = to_categorical(label_train, num_classes=2)
valY = to_categorical(label_val, num_classes=2)

aug = ImageDataGenerator(rotation_range=30, width_shift_range=0.1,
	height_shift_range=0.1, shear_range=0.2, zoom_range=0.2,
	horizontal_flip=True, fill_mode="nearest")

model = LeNet.build(width=55, height=55, depth=3, classes=2)
opt = Adam(lr=INIT_LR, decay=INIT_LR / EPOCHS)
model.compile(loss="binary_crossentropy", optimizer=opt,
	metrics=["accuracy"])

es = EarlyStopping(monitor='val_acc', mode='min', verbose=1, patience=20)

H = model.fit_generator(aug.flow(trainX, trainY, batch_size=BS),
 	validation_data=(valX, valY), steps_per_epoch=len(trainX) // BS,
 	epochs=EPOCHS, verbose=1,callbacks=[es])
```

```python
# save model
```

```python
save_model(model,'model.h5')
```

```python
# Generate plot of loss curve of train and validation
```

```python
plt.style.use("ggplot")
plt.figure()
N = len(H.epoch)
plt.plot(np.arange(0, N), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, N), H.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, N), H.history["acc"], label="train_acc")
plt.plot(np.arange(0, N), H.history["val_acc"], label="val_acc")
plt.title("Training Loss and Accuracy on Santa/Not Santa")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower left")
plt.savefig('_new_model.png')
```
